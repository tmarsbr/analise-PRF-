## Análise dos Acidentes nas Rodovias Brasileiras

### Objetivo da Análise
Este projeto visa realizar uma análise abrangente das causas dos acidentes nas rodovias do Brasil, focando especialmente na relação entre estes acidentes e o posicionamento dos radares. Através de uma metodologia robusta, combinaremos análise estatística e geoespacial para explorar vários aspectos:

#### Eficácia dos Radares
Investigação sobre a correlação entre a localização dos radares e a ocorrência de acidentes, avaliando a eficácia estratégica dos radares.

#### Concentração de Acidentes e Presença de Radares
Identificação das rodovias com maior número de acidentes e avaliação da cobertura de radares nessas áreas.

#### Gravidade dos Acidentes
Análise da severidade dos acidentes em diferentes trechos das rodovias e influência dos radares.

#### Análise Estatística de Correlação
Uso de técnicas estatísticas para determinar a relação entre a presença dos radares e a redução na frequência e gravidade dos acidentes.

#### Tipos de Acidentes ao Longo dos Anos
Exame das tendências nos tipos de acidentes ocorridos nas rodovias de 2022 a 2024.

#### Análise Temporal
Análise detalhada do período de três anos para identificar tendências, mudanças significativas e padrões emergentes.

---------------------------------------------------------------------------------------------
### Conclusão

Após uma análise detalhada dos dados dos acidentes nas rodovias brasileiras e a correlação destes com a localização dos radares, chegamos a várias conclusões importantes.

Primeiramente, identificamos que, apesar da presença dos radares em muitas áreas, ainda ocorrem acidentes significativos. Isso sugere que, além da fiscalização eletrônica, são necessárias medidas complementares de segurança, como melhorias na infraestrutura das rodovias, campanhas de conscientização para motoristas e reforço na sinalização de trânsito.

Observamos também que em alguns trechos com alta incidência de acidentes, a cobertura de radares é insuficiente. Isso indica a necessidade de uma avaliação estratégica sobre a localização dos radares, garantindo que eles estejam posicionados nos pontos mais críticos.

A análise da gravidade dos acidentes revelou que, em certos locais, apesar da presença de radares, os acidentes resultaram em consequências mais severas. Isso pode estar relacionado a fatores como condições climáticas adversas, características específicas da rodovia ou comportamentos de risco dos motoristas.

Por fim, a análise temporal mostrou que, embora algumas melhorias tenham sido observadas ao longo dos anos, ainda existem áreas que requerem atenção imediata para prevenir acidentes futuros. O declínio na frequência e gravidade dos acidentes em alguns trechos reforça a eficácia de medidas integradas de segurança, incluindo a presença de radares.

Em resumo, este estudo destacou a complexidade dos fatores que contribuem para os acidentes nas rodovias e a necessidade de uma abordagem multifacetada para melhorar a segurança. As recomendações apresentadas buscam oferecer um caminho para reduzir os acidentes e salvar vidas, contribuindo para estradas mais seguras em todo o Brasil.

8. **Utilização de Tecnologia Avançada**: Explorar o uso de tecnologias emergentes, como sistemas inteligentes de transporte e análise de dados em tempo real, para melhorar a prevenção e resposta a acidentes nas rodovias.

Estas recomendações visam criar um ambiente rodoviário mais seguro e reduzir a frequência e gravidade dos acidentes nas rodovias do Brasil. A implementação dessas estratégias requer uma abordagem colaborativa e o comprometimento de múltiplas partes interessadas, incluindo autoridades de trânsito, legisladores, comunidades locais e os próprios motoristas.

```python
def generate_metadata(dataframe):
    """
    Gera um dataframe contendo metadados das colunas do dataframe fornecido.

    :param dataframe: DataFrame para o qual os metadados serão gerados.
    :return: DataFrame contendo metadados.
    """

    # Coleta de metadados básicos
    metadata = pd.DataFrame({
        'nome_variavel': dataframe.columns,
        'tipo': dataframe.dtypes,
        'qt_nulos': dataframe.isnull().sum(),
        'percent_nulos': round((dataframe.isnull().sum() / len(dataframe))* 100,2),
        'cardinalidade': dataframe.nunique(),
    })
    metadata=metadata.sort_values(by='tipo')
    metadata = metadata.reset_index(drop=True)

    return metadata
generate_metadata(df_prf_oco_01)    
```

```python
# Convertendo a coluna 'km' para formato numérico, lidando com valores não numéricos
df_prf_oco_01['km'] = pd.to_numeric(df_prf_oco_01['km'], errors='coerce')

# Definindo um intervalo de quilômetros para considerar como 'próximo' a um radar
intervalo_proximidade = 50  # Por exemplo, 5 km

# Criando uma lista de intervalos de quilometragem próximos aos radares
intervalos_proximos_a_radares = [
    (radar - intervalo_proximidade, radar + intervalo_proximidade)
    for radar in df_radares['km_m'].unique()
]

# Função para verificar se um acidente ocorreu próximo a um radar
def proximo_a_radar(km, intervalos):
    if pd.isna(km):  # Verifica se o valor de 'km' é NaN
        return False
    for inicio, fim in intervalos:
        if inicio <= km <= fim:
            return True
    return False

# Aplicando a função aos dados de acidentes
df_prf_oco_01['proximo_a_radar'] = df_prf_oco_01.apply(lambda row: proximo_a_radar(row['km'], intervalos_proximos_a_radares), axis=1)

# Realizando as comparações e cálculos estatísticos
acidentes_proximos_a_radares = df_prf_oco_01[df_prf_oco_01['proximo_a_radar']]
acidentes_longe_de_radares = df_prf_oco_01[~df_prf_oco_01['proximo_a_radar']]

# Recalculando as estatísticas para comparação
estatisticas_proximos = acidentes_proximos_a_radares['classificacao_acidente'].value_counts(normalize=True)
estatisticas_longe = acidentes_longe_de_radares['classificacao_acidente'].value_counts(normalize=True)

# Exibindo as estatísticas corrigidas
estatisticas_proximos, estatisticas_longe




```

##### Acidentes Longe de Radares

##### Análise dos Resultados

   - Há uma **diminuição na proporção de acidentes fatais** perto dos radares (5,54%) em comparação com áreas mais distantes dos radares (9,59%). Isso pode indicar que os radares estão contribuindo para **reduzir a gravidade** dos acidentes.

2. **Impacto dos Radares na Frequência e Natureza dos Acidentes**
   - Os acidentes próximos aos radares tendem a ter uma proporção maior de **vítimas feridas** e uma menor proporção de acidentes **sem vítimas**. Isso sugere que, embora os radares possam estar ajudando a reduzir a letalidade, eles não estão necessariamente diminuindo a **frequência de acidentes que resultam em ferimentos**.


3. **Conclusão Geral sobre o Impacto dos Radares**
   - A presença de radares pode estar influenciando positivamente a segurança nas rodovias, **reduzindo a gravidade dos acidentes**, embora não necessariamente a **frequência total de ocorrências**.

```python
def plot_categorical_frequency_pt(df, corte_cardinalidade=30, graficos_por_linha=2):
    """
    Plota a frequência de categorias para variáveis categóricas em um DataFrame.

    Parâmetros:
    - df: DataFrame para plotagem.
    - corte_cardinalidade: Cardinalidade máxima para uma coluna ser considerada (padrão é 30).
    - graficos_por_linha: Quantidade de gráficos por linha (padrão é 2).

    Retorna:
    - Exibe os gráficos de barras.
    """

    # Gera metadados para o DataFrame
    metadados = []
    for coluna in df.columns:
        metadados.append({
            'Variável': coluna,
            'Tipo': df[coluna].dtype,
            'Cardinalidade': df[coluna].nunique()
        })

    df_metadados = pd.DataFrame(metadados)

    # Filtra colunas com cardinalidade menor ou igual ao corte e tipos não numéricos
    variaveis_categoricas = df_metadados[(df_metadados['Cardinalidade'] <= corte_cardinalidade) & (df_metadados['Tipo'] == 'object')]

    # Calcula o número de linhas e colunas para os subplots
    n_linhas = -(-len(variaveis_categoricas) // graficos_por_linha)  # Ceiling division
    n_colunas = min(len(variaveis_categoricas), graficos_por_linha)

    # Plota as variáveis categóricas
    fig, axs = plt.subplots(nrows=n_linhas, ncols=n_colunas, figsize=(15, 5 * n_linhas))

    for i, (idx, linha) in enumerate(variaveis_categoricas.iterrows()):
        var = linha['Variável']
        ax = axs[i // graficos_por_linha, i % graficos_por_linha]
        df[var].value_counts().sort_values(ascending=False).iloc[::-1].plot(kind='barh', ax=ax, color='skyblue')
        ax.set_title(f'Frequência em {var}')
        ax.set_xlabel('Frequência')
        ax.set_ylabel(var)

    # Remove os eixos vazios, se houver
    for j in range(i + 1, n_linhas * n_colunas):
        axs[j // graficos_por_linha, j % graficos_por_linha].axis('off')

    plt.tight_layout()
    plt.show()

```
